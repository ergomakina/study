```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = 'vibration.csv'
```

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
print(df.head())
```

```python
X = df['x']
Y = df['y']
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```
