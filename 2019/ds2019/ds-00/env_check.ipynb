### Sample run  

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
csv_in = 'world_country_popM.csv'
geo = 'ne_110m_admin_0_countries.geojson'
```

```python
df_popM = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
```

```python
m = folium.Map(location=[48,-102], zoom_start=1)

folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.gu_a3',
    data=df_popM,
    columns=['Country Code', '2017'],
    fill_color='BuPu',
    legend_name='Population (M) 2017'
).add_to(m)

display(m)
```
