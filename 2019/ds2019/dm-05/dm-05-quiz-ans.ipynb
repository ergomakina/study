### Import libraries  

```python
import pandas as pd
import folium
```

### Parameters  

```python
csv_in = 'swimming_pool.csv'
geo = 'japan.json'
```

### Read CSV files  

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=4, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

#### 1. Make choropleth map of #Swimming_Pool  

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Pref', 'Swimming_Pool'],
    fill_color='BuPu',
    legend_name='#Swimming_Pool'
).add_to(m)

display(m)
```

#### 2. Make choropleth map of #Swimming_Pool per 1M population  

```python
df['SP_M'] = df['Swimming_Pool'] / df['Population(M)']
display(df.head())
```

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Pref', 'SP_M'],
    fill_color='BuPu',
    legend_name='#Swimming_Pool_per_1M_population'
).add_to(m)

display(m)
```
