### Import libraries  

```python
import pandas as pd
```

### Parameters

```python
csv_in = 'crime2006_2016_estat.csv'
csv_out = 'crime2006_2016-utf8.csv'

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

### Read CSV file for check  

```python
df = pd.read_csv(csv_in, delimiter='\n', dtype='object',
                 skiprows=None, header=None, encoding='shift-jis')
print(df.shape)
display(df.head(15))
```

### Re-read CSV: set delimiter as ',', skip 8 lines and read header  
Set dtype automatically  

```python
df = pd.read_csv(csv_in, delimiter=',',
                 skiprows=8, header=0, encoding='shift-jis')
print(df.shape)
display(df.head(5))
```

### Number of lines to be skipped should be 10?  

```python
df = pd.read_csv(csv_in, delimiter=',',
                 skiprows=10, header=0, encoding='shift-jis')
print(df.shape)
display(df.head(15))
```

### Select columns we need  

```python
df2 = df[['時間軸(年次)', '管区警察局', '検挙人員【人】']]
df2.columns = ['年', '都道府県', '検挙人数']
df2['年'] = df2['年'].str.replace('年','',regex=True)
print(df2.shape)
print(df2.info())
print(df2)
```

### Drop lines other than Prefs

```python
df2['都道府県'].str.match(r'^(.*管区|.*方面|全国)$')  # debug
```

```python
df2016_crim = df2.drop(df2.index[ df2['都道府県'].str.match(r'^(.*管区|.*方面|全国)$') ]).reset_index(drop=True)
print(df2016_crim.info())
display(df2016_crim.head())
```

### Write to CSV file  

```python
df2016_crim.to_csv(csv_out, index=False)
```
