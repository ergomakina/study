### Import libraries  

```python
import pandas as pd
```

### Parameters

```python
csv_in = 'pop_move_H27_H28_estat.csv' # from e-Stat
csv_out = 'pop_move_H27_H28-utf8.csv'

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

### Read CSV file for check  

```python
df = pd.read_csv(csv_in, delimiter='\n', dtype='object',
                 skiprows=None, header=None, encoding='shift-jis')
print(df.shape)
display(df.head(15))
```

### Re-read CSV: set delimiter as ',', skip 10 lines and read header  
Set dtype automatically  

```python
df = pd.read_csv(csv_in, delimiter=',',
                 skiprows=10, header=0, encoding='shift-jis')
print(df.shape)
display(df.head(5))
```

### Select columns we need  

```python
df_diff = df[['Unnamed: 3', '実数　他都道府県（他市町村）からの転入者.1', '実数　他都道府県（他市町村）への転出者.1']]
print(df_diff.shape)
print(df_diff.info())
print(df_diff)
```

### Drop top 3 lines, reset columns

```python
df_diff2 = df_diff.loc[3:].reset_index(drop=True)
df_diff2.columns = ['都道府県', '転入', '転出']
print(df_diff2.shape)
print(df_diff2.info())
print(df_diff2)
```

### Save to CSV file

```python
df_diff2.to_csv(csv_out, index=False)
```
